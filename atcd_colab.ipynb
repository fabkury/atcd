# atcd — Run the WHO ATC-DDD scraper in Google Colab

Source repo: <https://github.com/fabkury/atcd>

This notebook clones the latest version of [`atcd`](https://github.com/fabkury/atcd) and runs `atcd.R` against the WHO Collaborating Centre for Drug Statistics Methodology website (<https://atcddd.fhi.no/atc_ddd_index/>). When it finishes, the freshly scraped CSV is written under `atcd/output/` and listed at the bottom of this notebook for download.

**Runtime:** Default Python runtime is fine — `r-base` is installed below.

**Tip:** This notebook deliberately does **not** inline the contents of `atcd.R`. It always runs the script straight from the repo, so it stays in sync with `master`.

```python
# 1. Install R and the packages atcd.R needs.
!apt-get -qq install -y r-base > /dev/null
!Rscript -e 'install.packages(c("pacman","rvest","dplyr","readr","xml2"), repos="https://cran.rstudio.com/", quiet=TRUE)'
```

```python
# 2. Clone the repository (or pull, if already present) and run atcd.R.
import os, subprocess
if not os.path.isdir('atcd'):
    subprocess.run(['git', 'clone', '--depth=1', 'https://github.com/fabkury/atcd.git'], check=True)
else:
    subprocess.run(['git', '-C', 'atcd', 'pull', '--ff-only'], check=True)
subprocess.run(['Rscript', 'atcd.R'], cwd='atcd', check=True)
```

```python
# 3. List the freshly scraped CSV(s).
import os
for f in sorted(os.listdir('atcd/output')):
    if f.endswith('.csv'):
        path = f'atcd/output/{f}'
        print(path, '—', os.path.getsize(path), 'bytes')
```
